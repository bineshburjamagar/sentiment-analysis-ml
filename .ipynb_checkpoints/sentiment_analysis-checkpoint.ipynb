```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data
reddit_df = pd.read_csv("data/Reddit_Data.csv")
reddit_df.head()
```

```python
reddit_df.size
```

```python
reddit_df.info()
```

```python
reddit_df.isna().sum()
```

```python
reddit_df = reddit_df.dropna(subset="clean_comment")
```

```python
reddit_df.clean_comment[0]
```

```python
reddit_df.clean_comment.value_counts()
```

```python
# visualizing sentiment analysis
sns.countplot(data=reddit_df, x='category')
plt.title("Sentiment Distribution")
plt.show()
```

```python
# exploring sentiment base insights
from wordcloud import WordCloud

positive_text = " ".join(reddit_df[reddit_df['category'] == 1]['clean_comment'])
negative_text = " ".join(reddit_df[reddit_df['category'] == -1]['clean_comment'])

WordCloud(width=800, height=400).generate(positive_text).to_image()
```

```python
WordCloud(width=800, height=400).generate(negative_text).to_image()
```

```python
# training model
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

# Vectorize
vectorizer = TfidfVectorizer(max_features=5000)
X = vectorizer.fit_transform(reddit_df['clean_comment'])
y = reddit_df['category']

# Split & Train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)
model = LogisticRegression()
model.fit(X_train, y_train)

# Evaluate
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
# Confusion matrix

sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
# testing new text

def predict_sentiment(text):
    cleaned = text.lower()
    vec = vectorizer.transform([cleaned])
    pred = model.predict(vec)[0]
    return pred

# Example
predict_sentiment("I love my life!")  # Should return 1
```

```python
# saving model for future use
import joblib

joblib.dump(model, "reddit_sentiment_model.pkl")
joblib.dump(vectorizer, "reddit_vectorizer.pkl")
```
