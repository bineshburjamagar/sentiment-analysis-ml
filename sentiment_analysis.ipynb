```python
# visualizing sentiment analysis
sns.countplot(data=reddit_df, x='category')
plt.title("Sentiment Distribution")
plt.show()
```

```python
# exploring sentiment base insights
from wordcloud import WordCloud

positive_text = " ".join(reddit_df[reddit_df['category'] == 1]['clean_comment'])
negative_text = " ".join(reddit_df[reddit_df['category'] == -1]['clean_comment'])

WordCloud(width=800, height=400).generate(positive_text).to_image()
```
